```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
# 目标混合高斯分布的参数
weights = [0.3, 0.7]  # 混合权重
means = [4, 7]        # 均值
variances = [0.3, 2]  # 方差
std_devs = np.sqrt(variances)  # 标准差
```

```python
def target_distribution(x):
    """混合高斯分布的概率密度函数"""
    p = 0
    for w, mu, sigma in zip(weights, means, std_devs):
        p += w * norm.pdf(x, loc=mu, scale=sigma)
    return p

def potential_energy(x):
    """势能函数U(x)"""
    return -np.log(target_distribution(x))

def grad_potential_energy(x):
    """势能函数的梯度"""
    p = target_distribution(x)
    grad = 0
    for w, mu, sigma in zip(weights, means, std_devs):
        grad += w * norm.pdf(x, loc=mu, scale=sigma) * (-1) * (x - mu) / (sigma ** 2)
    return -grad / p
```

```python
def leapfrog(x, p, grad_U, step_size, n_steps):
    """Leapfrog 数值积分方法"""
    p -= 0.5 * step_size * grad_U(x)  
    for _ in range(n_steps):
        x += step_size * p            
        if _ != n_steps - 1:          
            p -= step_size * grad_U(x)
    p -= 0.5 * step_size * grad_U(x)  
    return x, -p  
```

```python
def hmc_sampling(initial_x, num_samples, step_size, n_steps):
    """HMC 采样方法"""
    samples = []
    x = initial_x
    for _ in range(num_samples):
        # 随机初始化动量
        p = np.random.normal(0, 1)
        current_x, current_p = x, p
        
        # 使用 Leapfrog 方法进行模拟
        proposed_x, proposed_p = leapfrog(current_x, current_p, grad_potential_energy, step_size, n_steps)
        
        # 计算接受概率
        current_U = potential_energy(current_x)
        current_K = 0.5 * current_p ** 2
        proposed_U = potential_energy(proposed_x)
        proposed_K = 0.5 * proposed_p ** 2
        acceptance_prob = np.exp((current_U - proposed_U) + (current_K - proposed_K))
        
        # 决定是否接受新状态
        if np.random.rand() < acceptance_prob:
            x = proposed_x
        samples.append(x)
    return np.array(samples)
```

```python
# 设置 HMC 参数
initial_x = 0          # 初始值
num_samples = 5000     # 采样数量
step_size = 0.1        # Leapfrog 的步长
n_steps = 10           # Leapfrog 的步数

# 执行 HMC 采样
samples = hmc_sampling(initial_x, num_samples, step_size, n_steps)
```

```python
# 绘制目标分布和采样分布
x = np.linspace(-2, 12, 1000)
target_pdf = target_distribution(x)

plt.figure(figsize=(10, 6))
plt.plot(x, target_pdf, label="Target Distribution", color='blue', linewidth=2)
plt.hist(samples, bins=50, density=True, alpha=0.5, color='orange', label="HMC Sampled Distribution")
plt.title("HMC Sampling vs Target Distribution")
plt.xlabel("x")
plt.ylabel("Probability Density")
plt.legend()
plt.show()

# 计算误差指标 - KL 散度
from scipy.stats import gaussian_kde

def kl_divergence(p_samples, q_pdf, x_range):
    """计算 KL 散度"""
    p_pdf = gaussian_kde(p_samples).evaluate(x_range)
    p_pdf /= np.sum(p_pdf)  
    q_pdf /= np.sum(q_pdf)  
    return np.sum(p_pdf * np.log(p_pdf / q_pdf))

kl_div = kl_divergence(samples, target_pdf, x)
print(f"KL 散度: {kl_div}")
```
